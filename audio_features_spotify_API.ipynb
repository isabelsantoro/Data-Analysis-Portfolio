```python
# Initialize lists to store data
song_names = []
album_names = []
popularity_scores = []
audio_features = []

# Iterate over the albums and get the popularity of each track
for album in albums['items']:
    album_id = album['id']
    tracks = sp.album_tracks(album_id)['items']
    for track in tracks:
        track_info = sp.track(track['id'])
        song_names.append(track_info['name'])
        album_names.append(album['name'])
        popularity_scores.append(track_info['popularity'])
        audio_feature = sp.audio_features(track['id'])[0]
        audio_features.append(audio_feature)

# Create a dataframe from the collected data
data = {
    'Song': song_names,
    'Album': album_names,
    'Popularity': popularity_scores
}
df = pd.DataFrame(data)

# Extract audio feature column names
audio_feature_columns = list(audio_features[0].keys())

# Create a dictionary to store the audio features for each song
audio_features_dict = {column: [] for column in audio_feature_columns}

# Extract specific audio features and append them to the corresponding lists in the dictionary
for features in audio_features:
    for column in audio_feature_columns:
        audio_features_dict[column].append(features[column])

# Assign the audio features from the dictionary to new columns in the dataframe
for column in audio_feature_columns:
    df[column] = audio_features_dict[column]

# Print the updated dataframe with audio features
print(df)
```

```python
# Renaming some albums
df['Album'] = df['Album'].replace('Red (Deluxe Edition)', 'Red')
df['Album'] = df['Album'].replace('Fearless (International Version)', 'Fearless')

# List of albums to keep
albums_to_keep = ["Midnights", "evermore", "1989", "Fearless", "folklore", "Lover", "Red", "reputation", "Speak Now", "Taylor Swift"]

# Drop the unwanted albums
df = df[df['Album'].isin(albums_to_keep)]
```
